# Introduction 
In this project we will be compare the classification models,

those classification models are 

1. Decision Tree Classifier
2. Random Forest Classifier
3. KNN Classifier
4. Logistic Regression
5. SVM Classifier

Problem statement: Create a classification model to predict the gender (male or
female) based on different acoustic parameters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading data

```python
df=pd.read_csv("voice.csv")
```

# Data insights and adding label to male and female

```python
a=df.label.value_counts()
print(a)
print("Total: ",df.label.count())
```

```python
df.corr() #correlation between the various input variables
```

# our label feature has 2 valuble which are male and female. these are string but we need integers for classification therefore ,we must convert them from object to integer.

```python
df.label=[1 if each=="female" else 0 for each in df.label]
```

```python
df.label
```

# spliting data set into training set and testing set

```python
y=df.label.values
x=df.drop(['label'],axis='columns')
from sklearn.model_selection import train_test_split #import class for spliting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
#Hfere the size is 20%
algo_names=[]
algo_scores=[]
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train.head()
```

```python
x_train.tail()
```

# Applying different models 

```python
from sklearn.tree import DecisionTreeClassifier
#import decision tree classifier
dec_tree= DecisionTreeClassifier(random_state=50)
dec_tree.fit(x_train,y_train)
print("Decision Tree classification score:",dec_tree.score(x_test,y_test))#to print the accuracy
algo_names.append("Decision Tree")
algo_scores.append(dec_tree.score(x_test,y_test))
p1=dec_tree.predict(x_test)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
#import random forest classifier
rand_forest= RandomForestClassifier(random_state=50)
rand_forest.fit(x_train,y_train)
print("Random Forest classifier score: ",rand_forest.score(x_test,y_test))
algo_names.append("Random forest")
algo_scores.append(rand_forest.score(x_test,y_test))
p2=rand_forest.predict(x_test)
```

Svm classifier

```python
from sklearn.svm import SVC
#import the svc class
svm=SVC(random_state=50)
svm.fit(x_train,y_train)
print("SVM classifier score is {}".format(svm.score(x_test,y_test)))
algo_names.append("SVM")
algo_scores.append(svm.score(x_test,y_test))
p3=svm.predict(x_test)
```

KNN classifier

```python


from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
#when n_neighbors = 1 we get the maximum accuracy of the KNN Classifier
knn.fit(x_train, y_train)
print("KNN Classifier score is {}".format(knn.score(x_test,y_test)))
algo_names.append("KNN")
algo_scores.append(knn.score(x_test,y_test))
p4=knn.predict(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
print("Logistic Regression score is {}".format(lr.score(x_test,y_test)))
algo_names.append("Logistic Regression")
algo_scores.append(lr.score(x_test,y_test))
p5=lr.predict(x_test)
```

# Comparing the classification score of different algorithms using Bar Graph

```python
plt.figure(figsize=(12,10))
plt.ylim([0.60,1])
plt.bar(algo_names,algo_scores,width=0.5,color=['green'])
plt.xlabel('Algorithm Name')
plt.ylabel('Algorithm score')
```

# Plotting roc curve for differnt models using roc_curve of sklearn.metrics

Decision Tree Roc Curve

```python
from sklearn.metrics import roc_curve,roc_auc_score,auc
b=[]
probs1=dec_tree.predict_proba(x_test)
preds1=probs1[:,1]
fpr1,tpr1,threshold1=roc_curve(y_test,preds1)
roc_auc1=auc(fpr1,tpr1)
plt.title('Recevier Operating Characteristic')
plt.plot(fpr1,tpr1,'b',label='AUC=%0.2f' %roc_auc1)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('True positive rate')
plt.xlabel('False positive rate')
plt.title("Decision Tree Roc curve")
plt.show()
b.append(roc_auc1)
print("DECISION TREE AUC :",roc_auc_score(y_test,preds1))
```

Random Forest Roc Curve

```python
probs2=rand_forest.predict_proba(x_test)
preds2=probs2[:,1]
fpr2,tpr2,threshold2=roc_curve(y_test,preds2)
roc_auc2=auc(fpr2,tpr2)
plt.title('Recevier Operating Characteristic')
plt.plot(fpr2,tpr2,'b',label='AUC=%0.2f' %roc_auc2)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('True positive rate')
plt.xlabel('False positive rate')
plt.title("Random Forest Roc curve")
plt.show()
b.append(roc_auc2)
print("RANDOM FOREST AUC :",roc_auc_score(y_test,preds2))
```

SVM Roc Curve

```python
fpr3,tpr3,threshold3=roc_curve(y_test,p3)
roc_auc3=auc(fpr3,tpr3)
plt.title('Recevier Operating Characteristic')
plt.plot(fpr3,tpr3,'b',label='AUC=%0.2f' %roc_auc3)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('True positive rate')
plt.xlabel('False positive rate')
plt.title("SVM Roc curve")
plt.show()
b.append(roc_auc3)
print("SVM AUC :",roc_auc_score(y_test,p3))
```

KNN Roc Curve

```python
probs4=knn.predict_proba(x_test)
preds4=probs4[:,1]
fpr4,tpr4,threshold4=roc_curve(y_test,preds4)
roc_auc4=auc(fpr4,tpr4)
plt.title('Recevier Operating Characteristic')
plt.plot(fpr4,tpr4,'b',label='AUC=%0.2f' %roc_auc4)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('True positive rate')
plt.xlabel('False positive rate')
plt.title("KNN Roc curve")
plt.show()
b.append(roc_auc4)
print("KNN AUC :",roc_auc_score(y_test,preds4))
```
